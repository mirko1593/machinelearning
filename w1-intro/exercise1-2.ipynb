```python
import numpy as np
```

```python
x_train = np.array([1., 2.])
y_train = np.array([300., 500.])
print(f"x_train is {x_train} and shape is {x_train.shape}")
print(f"y_train is {y_train} with shape is {y_train.shape}")

m = x_train.shape[0]

print(f"no. of training examples: {m}")

m = len(x_train)
print(f"len of x_train: {m}")
```

```python
import matplotlib.pyplot as plt
plt.style.use("./deeplearning.mplstyle")
```

```python
plt.scatter(x_train, y_train, marker='x', c = 'r')

plt.ylabel("Price")
plt.xlabel("number of rooms")
plt.title("Hourse Price Prediction")

plt.show()
```

```python
w = 100
b = 100

print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = x[i] * w + b
        
    return f_wb
```

```python
w = 200
y_wb = compute_model_output(x_train, w, b)

print(y_wb)
```

```python
plt.plot(x_train, y_wb, c = 'b', label = 'our prediction')

plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')

plt.title("Hourse Price Prediction")

plt.xlabel("number of hourse")

plt.ylabel("price of hourse")

plt.legend()

plt.show()
```
